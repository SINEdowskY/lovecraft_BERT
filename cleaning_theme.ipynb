```python
import re
import os
import string
```

```python
TEXTS_DIR = r"lovecraft_novels"
DIR_LIST = os.listdir(TEXTS_DIR)
DIR_LIST
```

```python
with open(TEXTS_DIR+f"/{DIR_LIST[7]}") as file:
    lines = file.readlines()
    print(lines)
```

```python
lines[0] = lines[0].removeprefix("\ufeff")
```

## Lowercase

```python
lines = list(map(str.lower, lines))
lines
```

## Casefold

```python
lines = list(map(str.casefold, lines))
lines
```

## Removing empty lines

```python
lines = [line for line in lines if line != "\n"]
lines
```

## removing  \n \t etc.

```python
lines = list(map(lambda x: " ".join(x.split()), lines))
lines
```

## divide words
some words are connected like far-away so we need to change it into 2 words far away

```python
words = []
for line in lines:
    splited = line.split(" ")
    for word in splited:
        matches = re.findall(r'\b\w+(?:[-]{1,2}\w+)+\b', word)
        if len(matches) > 0:
            for match in matches:
                splited_dash = None
                if "--" in match:
                    splited_dash = match.split("--")
                else:
                    splited_dash = match.split("-")
                for sd in splited_dash:
                    words.append(sd)
        else:
            words.append(word)
```

```python
words
```

## Punctuation

```python
words = list(map(lambda x: re.sub(f"[{re.escape(string.punctuation)}]", "", x), words))
words
```

## Concat

```python
full_text = ""
for word in words:
    full_text += word+" "
full_text
```

```python
with open("clean_text.txt", mode="w+") as file:
    file.write(full_text)
    file.close()
```
