```python
import pandas as pd
import re
import os
```

```python
PATH="./lovecraft_novels/"
directory = os.listdir(path=PATH)
directory
```

```python
with open(PATH+directory[2], mode="r") as file:
    lines = file.readlines()
    rm_newline = [line.removesuffix("\n").removeprefix("\ufeff").lower() for line in lines]
    rm_whitespaces = []
    for line in rm_newline:
        rm_whitespaces.extend(line.split(" "))
    splited_text = [line for line in rm_whitespaces if line != "" and not bool(re.match(r'^-?\d+\.?\d*$', line))]
    
```

```python
for text in splited_text:
    if "word" in text:
        print(repr(text))
```

```python
clean_splited = []
for word in splited_text:
    clean = re.sub(r'[\ufeff;.,_"():?!]', '', word)
    if "-" in clean:
        splited_clean = clean.split("-")
        for splited in splited_clean:
            clean_splited.append(splited)
    elif "--" in clean:
        splited_clean = clean.split("--")
        for splited in splited_clean:
            clean_splited.append(splited)
    elif "\n" in clean:
        splited_clean = clean.split("\n")
        for splited in splited_clean:
            clean_splited.append(splited)
    else:
        clean_splited.append(clean)
    
```

```python
df = pd.DataFrame(
    data = {
        "words":clean_splited
    }
)
```
