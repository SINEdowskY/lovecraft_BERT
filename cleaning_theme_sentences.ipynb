```python
import re
import os
import string
import nltk
```

```python
TEXTS_DIR = r"lovecraft_novels"
DIR_LIST = os.listdir(TEXTS_DIR)
DIR_LIST
```

```python
with open(TEXTS_DIR+f"/{DIR_LIST[7]}") as file:
    lines = file.readlines()
    print(lines)
```

```python
lines[0] = lines[0].removeprefix("\ufeff")
lines = list(map(lambda x: " ".join(x.split()), lines))
lines
```

```python
one_line = ""
for line in lines:
    one_line += line+" "
one_line = one_line.replace("--", " ")
```

```python
nltk.download('punkt')
```

```python
def split_into_sentences(text):
    sentences = nltk.sent_tokenize(text)
    return sentences

sentences = split_into_sentences(one_line)
```

```python
for sentence in sentences:
    print(sentence)
```
